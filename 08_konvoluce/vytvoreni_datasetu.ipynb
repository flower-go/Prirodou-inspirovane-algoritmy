```python
def png_to_jpg(file):
    im = Image.open(file)
    bg = Image.new("RGB", im.size, (255,255,255))
    bg.paste(im,im)
    bg.save(file[0:-3] + "jpg")
```

```python
mypath = '/Users/petravysusilova/Documents/school/prirodouInspAlg/dataset_hp'
directories = ["dumbledore","voldemort","mcgonagal","snape"]
all_all = []
for d in directories:
    path = mypath + "/" + d
    all_images = [f for f in listdir(path) if isfile(join(path, f)) and f[0] != "." and not (f[0] == "s" and f[1] == "_" ) ]
    all_all.append(all_images)
#print(all_all)
```

```python
for all_images,d in zip(all_all,directories):
    for f in all_images:
        if f[-3:] == "png":
            png_to_jpg(mypath + "/" + d + "/" + f)
            
```
